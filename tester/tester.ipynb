```python
from datetime import datetime
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
import labwons as lw
```

```python
equity = lw.Equity('SPY')
```

```python
equity.trend

fig = make_subplots(
    rows=2, cols=3,
)

traces = list()
for col in equity.trend.columns:
    model = pd.concat(objs=[equity.trend[col], equity.typical], axis=1)
    model = model.dropna()
    model[f'Diff-{col}'] = 100 * (model[model.columns[1]] / model[model.columns[0]] - 1)
    traces.append(
        go.Scatter(
            name=f'Diff-{col}',
            x=model.index,
            y=model[f'Diff-{col}'],
            meta=model[model.columns[1]],
            text=model[model.columns[0]],
            hovertemplate='%{y:.2f}<br>' + model.columns[1] + ': %{meta:.2f}<br>' + model.columns[0] + ': %{text}<extra></extra>'
        )
    )

# model = pd.concat(objs=[equity.trend['TL(H)'], equity.typical], axis=1)
# model = model.dropna()
# model['Diff'] = 100 * (model[model.columns[1]] / model[model.columns[0]] - 1)
# # _1m['Diff'] = _1m[_1m.columns[1]] - _1m[_1m.columns[0]]
# model = model[model.index >= datetime(2010, 5, 1).date()]
# fig = go.Figure(
#     data=go.Scatter(
#         name='Diff',
#         x=model.index,
#         y=model['Diff'],
#         meta=model[model.columns[1]],
#         text=model[model.columns[0]],
#         hovertemplate='%{y:.2f}<br>' + model.columns[1] + ': %{meta:.2f}<br>' + model.columns[0] + ': %{text}<extra></extra>'
#     )
# )
# fig.show()
```
